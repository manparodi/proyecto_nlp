```python
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx])
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:15]
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import defaultdict

# Define a function to tokenize and count sentences in the text
def count_sentences(text):
    sentence_list = sent_tokenize(text)
    sentence_dict = defaultdict(int)
    for sentence in sentence_list:
        sentence_dict[sentence] += 1
    return sentence_dict

# Concatenate all message bodies into a single string
text = ''
for messages in conversations.values():
    for message in messages:
        text += message['body']

# Tokenize and count sentences
sentence_dict = count_sentences(text)

# Sort the sentences by count and print the top 10 most recurrent ones
sorted_sentences = sorted(sentence_dict.items(), key=lambda x: x[1], reverse=True)
for sentence, count in sorted_sentences[:5]:
    print(f'Sentence: {sentence.strip()}')
    print(f'Count: {count}')
    print('\n')
```
